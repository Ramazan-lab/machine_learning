```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/column_2C_weka (1).csv")

# %%
df.head(20)
```

```python
plt.scatter(df.pelvic_incidence,df.lumbar_lordosis_angle)
plt.xlabel("pelvic_incidence")
plt.ylabel("lumbar_lordosis_angle")
plt.show()
```

```python
df.rename(
    columns=({ 'class': 'pclass'}), 
    inplace=True,
)
df.tail()
```

```python
df.pclass = [1 if each == "Abnormal" else 0 for each in df.pclass]
y = df.pclass.values
x_data = df.drop(["pclass"],axis=1)
df.head()
```

```python
from sklearn.linear_model import LinearRegression

# linear regression model
linear_reg = LinearRegression()

x = df.pclass.values.reshape(310,1)
y = df.pelvic_incidence.values.reshape(310,1)

linear_reg.fit(x,y)
x,y
```

```python
import numpy as np

b0 = linear_reg.predict([[0]])
print("b0: ",b0)
```

```python
b0_ = linear_reg.intercept_
print("b0_: ",b0_)   # intercept the y-axis
```

```python
b11 = linear_reg.predict([[11]])
print("b11: ",b11)
```
